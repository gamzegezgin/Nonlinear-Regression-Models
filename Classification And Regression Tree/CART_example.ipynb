**Amaç:** Veri setindeki karmaşık yapıları basitleştirmektir.

Heterojen veri setleri(veri setinin ilk hali) belirlenmiş bir hedef değişkene(bağımlı değişken) göre homojen alt gruplara ayrılır.

Çok iyi sonuçlar vermeyebilir fakat karar ağaçlarının temelini oluşturur.

Genellemeye kaygısı duymadan mevcut durumu ifade etmek için ML algoritması kullandığımız senaryolarda en başarılı algoritmalardan birisidir.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
df = pd.read_csv("../Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League","Division", "NewLeague"]])
y = df["Salary"]
X_ = df.drop(["League", "Division" , "NewLeague" ,"Salary"], axis=1).astype("float64")
X = pd.concat([X_, dms[["League_N","Division_W", "NewLeague_N"]]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=33)
```

```python
X_train = pd.DataFrame(X_train["Hits"])
X_test = pd.DataFrame(X_test["Hits"])
```

```python
X_test
```

```python
cart_model = DecisionTreeRegressor(max_leaf_nodes = 10)
```

```python
cart_model.fit(X_train, y_train)
```

```python
X_grid = np.arange(min(np.array(X_train)), max(np.array(X_train)), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X_train, y_train, color='red')
plt.plot(X_grid, cart_model.predict(X_grid), color='blue')

plt.title('CART REGRESON AĞACI')
plt.xlabel('Atış Sayısı (Hits)')
plt.ylabel('Maaş (Salary)')
```

Dallar atış sayısının >120, maaş ise >450 şeklinde ayrılacak.(grafiğe göre yaklaşık degerler yazdim.)

leaf_nodes artıkca oluşan alanların içerisindeki gözlemler azalır ve tahminin daha kolay hale gelmesi beklenir.

```python
#tek değişkenli tahmin
y_pred = cart_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
df = pd.read_csv("../Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League","Division", "NewLeague"]])
y = df["Salary"]
X_ = df.drop(["League", "Division" , "NewLeague" ,"Salary"], axis=1).astype("float64")
X = pd.concat([X_, dms[["League_N","Division_W", "NewLeague_N"]]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=33)
```

```python
cart_model = DecisionTreeRegressor().fit(X_train, y_train)
```

```python
y_pred = cart_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

## **Model Tuning**

```python
cart_model = DecisionTreeRegressor(max_depth = 4).fit(X_train, y_train)
y_pred = cart_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# max_depth --> agacin ne kadar derinlesebilecegini ifade eder
# max_leaf_nodes --> dallanma kontrolü icin
# min_samples_split --> dallanma kontrolü icin
# min_samples_leaf
# bu hiperparametrelerin optimum değerlenini arıyoruz
# if-else durumunda öteye götürmek 
```

```python
cart_cv_model = GridSearchCV(cart_model, cart_params, cv=10).fit(X_train, y_train)
```

```python
#final
cart_tuned= DecisionTreeRegressor(max_depth=5, min_samples_split=50).fit(X_train, y_train)
```

```python
y_pred = cart_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```
